## Import:

```python
!pip install -e ..
import nextbike
import folium
import numpy as np
from folium.plugins import HeatMap, HeatMapWithTime


```

## Read the data:


```python
dfRaw = nextbike.io.read_file()
```

```python
dfStationBikeNumber = nextbike.io.preprocessStationData(dfRaw)
dfStationBikeNumber = nextbike.io.createBikeNumberPerStationIndex(dfStationBikeNumber)
dfTrips = nextbike.io.preprocessData(dfRaw)
dfStations = nextbike.io.createStations(dfTrips)
dfTrips = nextbike.io.createTrips(dfTrips)
dfStations
```

## Visualization:
### Creating the underlying map and marking the stations

```python
m = folium.Map(location=[50.8008, 8.7667], zoom_start=13, tiles='Stamen Toner')

for index, row in dfStations.iloc[1:].iterrows():     
        folium.CircleMarker(
            location=[row['pLat'], row['pLong']],
            popup=row['pName'],
            color='black',
            fill=True,
            fill_color='#3186cc'
        ).add_to(m)

f = folium.FeatureGroup(name='Start locations')
f2 = folium.FeatureGroup(name='End locations')
f3 = folium.FeatureGroup(name='Mixed weights')
f4 = folium.FeatureGroup(name='Mixed weights2')
```

### Creating the heatmap

```python
movingData = dfTrips.loc[dfTrips['sTime'].dt.date == np.datetime64('2019-08-01')]
heatData = np.c_[movingData[['sLat','sLong']].values.tolist(),np.ones(len(movingData['eLat']))*(1)].tolist()
heatData2 = np.c_[movingData[['eLat','eLong']].values.tolist(),np.ones(len(movingData['eLat']))*(1)].tolist()
heatData3 = np.r_[np.c_[movingData[['sLat','sLong']].values.tolist(),np.ones(len(movingData['eLat']))*(1)].tolist(),np.c_[movingData[['eLat','eLong']].values.tolist(),np.ones(len(movingData['eLat']))*(-1)].tolist()].tolist()
heatData4 = np.r_[np.c_[movingData[['sLat','sLong']].values.tolist(),np.ones(len(movingData['eLat']))*(-1)].tolist(),np.c_[movingData[['eLat','eLong']].values.tolist(),np.ones(len(movingData['eLat']))*(1)].tolist()].tolist()
#heatData3
```

```python
HeatMap(heatData, min_opacity=0.5,max_val=50,name='Start',radius=30).add_to(f)
HeatMap(heatData2, min_opacity=0.5,max_val=50,name='Start',radius=30).add_to(f2)
HeatMap(heatData3, min_opacity=0.5,max_val=50,name='Start',radius=30).add_to(f3)
HeatMap(heatData4, min_opacity=0.5,max_val=50,name='Start',radius=30).add_to(f4)

f.add_to(m)
f2.add_to(m)
f3.add_to(m)
f4.add_to(m)
folium.LayerControl().add_to(m)
m
```

```python
# timeframe is the amount of hours around the cutTime and date that are rendered. Date example '2019-08-01'
def createSplitHeatmap(dfTrips,dfStations,date,timeframe=12, startOrend = 'start'):
    if startOrend == 'start':
        Lat1 = 'sLat'
        Long1 = 'sLong'
        Lat2 = 'sLat'
        Long2 = 'sLong'
    elif startOrend == 'end':
        Lat1 = 'eLat'
        Long1 = 'eLong'
        Lat2 = 'eLat'
        Long2 = 'eLong'
    else:
        Lat1 = 'eLat'
        Long1 = 'eLong'
        Lat2 = 'sLat'
        Long2 = 'sLong'
    # Prepare the data
    movingDataS = dfTrips.loc[(dfTrips['sTime'] >= (np.datetime64(date) - np.timedelta64(timeframe,'h'))) & (dfTrips['sTime'] <= (np.datetime64(date)))]
    movingDataE = dfTrips.loc[(dfTrips['sTime'] >= (np.datetime64(date) )) & (dfTrips['sTime'] <= (np.datetime64(date) + np.timedelta64(timeframe,'h')))]
    

    heatDataS = np.c_[movingDataS[[Lat1,Long1]].values.tolist(),np.ones(len(movingDataS[Lat1]))*(1)].tolist()
    heatDataE = np.c_[movingDataE[[Lat2,Long2]].values.tolist(),np.ones(len(movingDataE[Lat2]))*(1)].tolist()
    heatDataSE = np.r_[np.c_[movingDataS[[Lat1,Long1]].values.tolist(),np.ones(len(movingDataS[Lat1]))*(1)].tolist(),np.c_[movingDataE[[Lat2,Long2]].values.tolist(),np.ones(len(movingDataE[Lat2]))*(-1)].tolist()].tolist()

    # Create the map
    map = folium.Map(location=[50.8008, 8.7667], zoom_start=13, tiles='Stamen Toner')
    # Add stations as black circle
    for index, row in dfStations.iloc[1:].iterrows():     
        folium.CircleMarker(
            location=[row['pLat'], row['pLong']],
            popup=row['pName'],
            color='black',
            fill=True,
            fill_color='#3186cc'
        ).add_to(map)

    f = folium.FeatureGroup(name='Start locations early')
    f2 = folium.FeatureGroup(name='Start locations later')
    f3 = folium.FeatureGroup(name='Difference in start locations.')
    HeatMap(heatDataS, min_opacity=0.4,max_val=25,name='Start',radius=30).add_to(f)
    HeatMap(heatDataE, min_opacity=0.4,max_val=25,name='End',radius=30).add_to(f2)
    HeatMap(heatDataSE, min_opacity=0.4,max_val=25,name='End2',radius=30,show=False).add_to(f3)
    f.add_to(map)
    f2.add_to(map)
    f3.add_to(map)
    folium.LayerControl().add_to(map)
    return map


map = createSplitHeatmap(dfTrips,dfStations,'2019-10-09 13:00',startOrend = 'end', timeframe=5)
map
```
